### Supply Graphs

```python
import sc2reader
# from sc2reader.factories.plugins.utils import PlayerSelection
# from sc2reader.factories.plugins.utils import GameState
from sc2reader.engine.plugins import ContextLoader, APMTracker
from sc2reader.engine.plugins import SelectionTracker
import time
```

```python
class Plugin1():
    def __init__(self):
        self.name = "Plugin1"
        self.player_1_frames = []
        self.player_2_frames = []
        self.player_1_supply = []
        self.player_2_supply = []

    def handlePlayerStatsEvent(self, event, replay):
        if event.pid == 1:
            self.player_1_frames.append(event.frame)
            self.player_1_supply.append(event.food_used)
        elif event.pid == 2:
            self.player_2_frames.append(event.frame)
            self.player_2_supply.append(event.food_used)


```

```python
plugin1 = Plugin1()

replay = sc2reader.load_replay(
    'thereplay.SC2Replay',
    engine=sc2reader.engine.GameEngine(plugins=[
        ContextLoader(),
        APMTracker(),
        plugin1,
        SelectionTracker(),
    ])
)

```
